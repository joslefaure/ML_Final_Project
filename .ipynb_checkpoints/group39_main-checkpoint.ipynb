```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
import numpy as np

# read my data
df1 = pd.read_csv("scraping/AMD/2018.12.04__2015.12.04/AMD.csv")
df2 = pd.read_csv("scraping/AVGO/2018.12.04__2015.12.04/AVGO.csv")
df3 = pd.read_csv("scraping/MU/2018.12.04__2015.12.04/MU.csv")
df4 = pd.read_csv("scraping/PLAB/2018.12.04__2015.12.04/PLAB.csv")
df5 = pd.read_csv("scraping/TSM/2018.12.04__2015.12.04/TSM.csv")
df6 = pd.read_csv("scraping/XPER/2018.12.04__2015.12.04/XPER.csv")

# transform the datetime (optional)
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])
df3['Date'] = pd.to_datetime(df3['Date'])
df4['Date'] = pd.to_datetime(df4['Date'])
df5['Date'] = pd.to_datetime(df5['Date'])
df6['Date'] = pd.to_datetime(df6['Date'])

# drop the date (unused)
df1.drop(['Date', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)
df2.drop(['Date', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)
df3.drop(['Date', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)
df4.drop(['Date', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)
df5.drop(['Date', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)
df6.drop(['Date', 'High', 'Low', 'Adj Close'], axis=1, inplace=True)


# add new col with  day count (the earliest day being day 1 (2015-12-04))
# and the last (2018-12-03)
df1.insert(0, 'Day_count', range(len(df1), 0, -1))
df2.insert(0, 'Day_count', range(len(df2), 0, -1))
df3.insert(0, 'Day_count', range(len(df3), 0, -1))
df4.insert(0, 'Day_count', range(len(df4), 0, -1))
df5.insert(0, 'Day_count', range(len(df5), 0, -1))
df6.insert(0, 'Day_count', range(len(df6), 0, -1))

# preprocess the volume to  make the graph clearer 
# volume will be in millon unit
df1['Volume'] = df1['Volume']/1000000
df2['Volume'] = df2['Volume']/1000000
df3['Volume'] = df3['Volume']/1000000
df4['Volume'] = df4['Volume']/1000000
df5['Volume'] = df5['Volume']/1000000
df6['Volume'] = df6['Volume']/1000000

# separate the features from the output (same for all of the dataframes)
features1 = df1.drop("Day_count", axis=1)
features2 = df2.drop("Day_count", axis=1)
features3 = df3.drop("Day_count", axis=1)
features4 = df4.drop("Day_count", axis=1)
features5 = df5.drop("Day_count", axis=1)
features6 = df5.drop("Day_count", axis=1)



# print(df1)

# Graph for the first  company AMD
for column in features1:
    fig, ax = plt.subplots()
    plt.plot(features1[column], df1['Day_count'])
    ax.set_xlabel(column)
    ax.set_ylabel('Day_count')
    plt.show()
```

```python
# Apply linear regression to DF1
array = np.array(df1)
X = array[:,1].reshape(-1, 1)
Y = array[:,0]

# Linear regression
train_X, test_X, train_y, test_y = train_test_split(X, Y, test_size=0.20)
reg = LinearRegression().fit(train_X, train_y)
pred_y = reg.predict(test_X)

# accuracy
acc = reg.score(test_X, test_y)
print("accuracy: ", acc)

# Plot the graph
plt.scatter(test_X, test_y,  color='black', label = 'Scatter plots')
plt.plot(test_X, pred_y, color='red', linewidth=3, label='Regression Line')
plt.legend()
plt.show()
```
